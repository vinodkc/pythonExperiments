```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup

channelname = "IndianMominCalifornia"
youtubeurl  = f'https://www.youtube.com/@{channelname}/videos'
driver = webdriver.Chrome()
driver.get(youtubeurl)

from selenium.webdriver.common.by import By

for i in range(5):
     driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
     time.sleep(5)
     print(f'scolling : {i}')

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
videos = soup.find_all('div', {'id':'dismissible'})

master_list = []
for video in videos:
    data_dict = {}
    title = video.find('a', {'id':'video-title-link'})
    data_dict['title'] = title.text
    data_dict['video_url'] = "https://www.youtube.com/" + title['href']
    meta = video.find('div', {'class':'ytd-video-meta-block'}).find_all('span')
    data_dict['views'] = meta[0].text
    data_dict['age'] = meta[1].text
    master_list.append(data_dict)

import pandas as pd

youtube_df = pd.DataFrame(master_list)

def convert_views(df):
    views = df['views'].split(' ')[0]
    if 'K' in df['views']:
        return float(views.split('K')[0]) * 1000
        
    elif 'M' in df['views']:
        return float(views.split('M')[0]) * 1000000
    else:
        return float(views)

def convert_age(df):
    age = df['age'].split(' ')
    if 'years' == age[1] or 'year' == age[1]:
        return int(age[0]) * 365
    elif 'months' == age[1] or 'month' == age[1]:
        return int(age[0]) * 30  
    elif 'weeks' == age[1]:
        return int(age[0]) * 7
    else:
        return int(age[0]) 

def find_monitizingratio_ration(df):
    return float(df['views'] / df['age'])
        
         
youtube_df["views"]  = youtube_df.apply(convert_views, axis=1).astype(int)
youtube_df["age"]  = youtube_df.apply(convert_age, axis=1)
youtube_df["monitizingratio"]  = youtube_df.apply(find_monitizingratio_ration, axis=1)

youtube_df = youtube_df.sort_values(by=['monitizingratio'], ascending=False)

output_filename = f"/tmp/youtube_video_details_of_channel_{channelname}.csv"
youtube_df.to_csv(output_filename, index=False)
print(f'csv file written to {output_filename}')
youtube_df
```
